```python
import matplotlib.pyplot as plt

marks = [55,60,61,62,63,65,70,71,72,80,82,90]

plt.hist(marks,bins=5)

plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")

plt.show()
```

```python
hours = [1,2,3,4,5]
marks = [35,45,55,65,75]

plt.scatter(hours,marks)

plt.xlabel("Hours")
plt.ylabel("Marks")

plt.show()
```

```python
students = ["John","Alice","David"]
marks = [75,90,82]

plt.bar(students,marks)

plt.title("Student Marks")
plt.show()
```

```python
labels = ["Python","Java","C++"]

sizes = [50,30,20]

plt.pie(sizes,
        labels=labels,
        autopct="%1.1f%%")

plt.show()
```
